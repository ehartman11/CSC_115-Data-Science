```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Print the first 5 rows to observe your data
iris_df.head()
```

```python

# View dataset info
iris_df.info()
```

```python
# Compute mean, median, and mode for each feature
median = iris_df.median()
mean = iris_df.mean()
mode = iris_df.mode()
("Median", median), ("Mean", mean), ("Mode", mode)
```

```python
# Calculate the range, variance and standard deviation for each feature
column_range = iris_df.max() - iris_df.min()
variance = iris_df.var()
std_dev = iris_df.std()
("Range", column_range), ("Variance", variance), ("Standard Deviation", std_dev)
```

```python
# Compute the 25th, 50th, and 75th, percentiles (quartiles) for each feature
quartiles = iris_df.quantile([.25, .50, .75])
quartiles
```

```python
# Create a histogram for each of the features to visualize their distributions
for col in iris_df.columns:
    plt.figure(figsize=[8, 5])
    sns.histplot(data=iris_df, x=col)
    plt.show()
```

```python
# Plot scatter plots between pairs of features to explore relationships
for col in iris_df.columns:
    for i in range(3):
        if col != iris_df.columns[i]:
            plt.figure(figsize=[8, 5])
            plt.scatter(data=iris_df, x=col, y=iris_df.columns[i])
            plt.xlabel(col)
            plt.ylabel(iris_df.columns[i])
            plt.show()
```
